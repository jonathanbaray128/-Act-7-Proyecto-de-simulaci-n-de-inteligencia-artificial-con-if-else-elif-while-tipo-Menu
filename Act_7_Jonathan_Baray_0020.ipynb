### Proyecto de simulación de inteligencia artificial con if, else, elif  while tipo Menu
#### Contador

```python
# ============================================
# SISTEMA INTELIGENTE DE CONTADOR Y CAJA
# MENU INTERACTIVO CON WHILE TRUE
# ============================================

# VARIABLES GENERALES
subtotal = 0
descuento = 0
comision = 0
total = 0
tipo_cliente = 0
metodo_pago = 0

# CICLO PRINCIPAL
while True:

    print("\n===================================")
    print("  SISTEMA INTELIGENTE CONTADOR")
    print("            Y CAJA")
    print("===================================")

    print("1. Registrar compra")
    print("2. Ver análisis inteligente")
    print("3. Salir")

    opcion = int(input("Seleccione una opción: "))

    # ============================================
    # OPCION 1
    # ============================================

    if opcion == 1:

        print("\nTIPO DE CLIENTE")
        print("1. Cliente normal")
        print("2. Cliente frecuente")
        print("3. Cliente VIP")

        tipo_cliente = int(input("Opción: "))

        print("\nMÉTODO DE PAGO")
        print("1. Efectivo")
        print("2. Tarjeta")
        print("3. Transferencia")

        metodo_pago = int(input("Opción: "))

        subtotal = float(input("\nIngrese el subtotal de compra: "))

        # DESCUENTOS
        if tipo_cliente == 1:
            descuento = subtotal * 0.05
            mensaje_cliente = "Cliente normal con descuento del 5%"

        elif tipo_cliente == 2:
            descuento = subtotal * 0.10
            mensaje_cliente = "Cliente frecuente con descuento del 10%"

        elif tipo_cliente == 3:
            descuento = subtotal * 0.20
            mensaje_cliente = "Cliente VIP con descuento del 20%"

        else:
            descuento = 0
            mensaje_cliente = "Cliente no válido"

        # COMISIONES
        if metodo_pago == 1:
            comision = 0
            mensaje_pago = "Pago en efectivo sin comisión"

        elif metodo_pago == 2:
            comision = subtotal * 0.03
            mensaje_pago = "Pago con tarjeta con comisión del 3%"

        elif metodo_pago == 3:
            comision = subtotal * 0.01
            mensaje_pago = "Pago por transferencia con comisión del 1%"

        else:
            comision = 0
            mensaje_pago = "Método de pago inválido"

        # TOTAL
        total = subtotal - descuento + comision

        print("\n===================================")
        print("      COMPRA REGISTRADA")
        print("===================================")

        print("Subtotal: $", subtotal)
        print("Descuento: $", descuento)
        print("Comisión: $", comision)
        print("Total a pagar: $", total)

    # ============================================
    # OPCION 2
    # ============================================

    elif opcion == 2:

        print("\n===================================")
        print("      ANÁLISIS INTELIGENTE")
        print("===================================")

        print(mensaje_cliente)
        print(mensaje_pago)

        # NIVEL DEL CLIENTE
        if total >= 10000:
            nivel = "CLIENTE PREMIUM"

        elif total >= 5000:
            nivel = "CLIENTE IMPORTANTE"

        elif total >= 1000:
            nivel = "CLIENTE ESTÁNDAR"

        else:
            nivel = "CLIENTE BÁSICO"

        print("Nivel detectado:", nivel)

        # ALERTAS
        if total > 15000:
            print("\nALERTA:")
            print("Compra muy alta")
            print("El contador recomienda revisión fiscal")

        elif total > 7000:
            print("\nALERTA:")
            print("Validar autorización bancaria")

        else:
            print("\nSistema funcionando correctamente")

        if tipo_cliente == 3:
            print("Beneficio VIP aplicado correctamente")

    # ============================================
    # OPCION 3
    # ============================================

    elif opcion == 3:

        print("\nGracias por usar el sistema inteligente")
        print("Programa finalizado")
        break

    # ============================================
    # OPCION INVALIDA
    # ============================================

    else:
        print("\nOpción inválida")
        print("Intente nuevamente")
```
